```python
df_3=ratings.merge(books,on='ISBN')
```

```python
df_4
```

```python
user_count=df_4.groupby(['User-ID']).count()
```

```python
user_count1=user_count[user_count['Book-Title']>200]
```

```python
temp=df_4.groupby('Book-Title').count()
```

```python
ratings_count=temp[temp['User-ID']>50]
```

```python
fin1=df_4.merge(user_count1,on='User-ID',how='inner')
```

```python
fin2=fin1.drop(['Book-Title_y','Book-Rating_y'],axis=1).rename(columns={'Book-Title_x':'Book-Title','Book-Rating_x':'Book-Rating'})
```

```python
fin2
```

```python
fin3=fin2.merge(,on='Book-Title',how='inner').drop(['User-ID_y','Book-Rating_y'],axis=1).rename(columns={'User-ID_x':'User-ID','Book-Rating_x':'Book-Rating'})
```

```python
fin4=fin3.merge(df_3,on=['Book-Title','User-ID'])
```

```python
fin4=fin4.rename({'Book-Rating_x':'Book-Rating'},axis=1)
```

```python
pt=fin4.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating')
```

```python
pt.fillna(0,inplace=True)
```

```python
pt
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(pt)
```

```python
similarity
```

```python
def recommend(name):
    try:
        name=str(name)
        idx=np.where(pt.index==name)[0][0]
        rec_idx=some(idx)
        for i in rec_idx:
            print(pt.index[i])
    except IndexError:
        print('Movie not available')
def some(idx):
    vector=similarity[idx]
    sort_vect=sorted(enumerate(vector),reverse=True,key=lambda x: x[1])
    return [ x[0] for x in sort_vect[1:6]]
```

```python
recommend('The Notebook')
```
